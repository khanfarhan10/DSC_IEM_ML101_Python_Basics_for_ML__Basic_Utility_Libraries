#### **Factorials, Permutations & Combinations**  

**Factorial :**
Product of all positive integers less than or equal to n :


$$
n !=n \cdot(n-1) \cdot(n-2) \cdot(n-3) \cdots \cdots \cdot 3 \cdot 2 \cdot 1
$$

````
print ("The factorial of 5 is :",math.factorial(5))
````

**Permutation :**

$$
P(n, k)=\underbrace{n \cdot(n-1) \cdot(n-2) \cdots(n-k+1)}_{k \text { factors }},
$$
which is 0 when $k>n,$ and otherwise is equal to
$$
^nP_k = P(n, k)=\frac{n !}{(n-k) !}



$$

````
def PermutationCoeff(n,k):
  if k>n:
    return 0
  return math.factorial(n)/math.factorial(n-k)
PermutationCoeff(5,2)
````

**Combination :**
$$
^nC_k = C (n,k)=\left(\begin{array}{l}
n \\
k
\end{array}\right)=\frac{n !}{k !(n-k) !}
$$

````
def CombinationCoeff(n,k):
  if k>n:
    return 0
  return math.factorial(n)/(math.factorial(n-k)*math.factorial(k))
CombinationCoeff(7,2)
````

**Euclidean Distance :**  
$$
d(\mathbf{p}, \mathbf{q})=\sqrt{\sum_{i=1}^{n}\left(q_{i}-p_{i}\right)^{2}}
$$

$$
\begin{aligned}
&p, q=\text {  two points in Euclidean n-space }\\
&\begin{array}{l}
q_{i}, p_{i}=\begin{array}{l}
\text { Euclidean vectors, starting from the origin of the space } \\
\text { (initial point) }
\end{array} \\
n \quad=n \text { -space }
\end{array}
\end{aligned}
$$
 
`math.dist(p, q)`

````
def EuclideanDist(p,q):
  dist=0
  for i in range(len(p)):
    dist=dist+math.pow((p[i]-q[i]),2)
  return math.sqrt(dist)
p=[5,3]
q=[1,6] 
EuclideanDist(p, q)
````

#### **Solve This :  Coding Challenge Problem**

Farhan is struggling with Cubic Polynomials with Real Roots. Try to write the following algorithm to find the roots of a cubic equation (or any eqn for that matter).

$$f(x) = x^3 - 6x-4 = 0$$

**Algorithm :**
- Initialize $x_0 =-4 (\text{A random guess value})$
- Initialize $\epsilon = 10^{-6} (\text{A very small number})$
- Repeat until convergence($|x_{n+1} -x_{n}|< \epsilon$) {
$$
x_{n+1}=x_{n}-\frac{f\left(x_{n}\right)}{f^{\prime}\left(x_{n}\right)}
$$}

````
def differentiate(func,a,h=1e-6):
  """
  Returns a derivative of the passed function f(x) at a given value a using the concept of Right Hand and Left Hand derivative approximation.

  f = univariate function f(x) 
  a = value at which derivative should be calculated f'(x=a)
  h = the tolerance h, which is a value very close to zero but not zero.

  For more details view :
  http://archives.math.utk.edu/visual.calculus/2/definition.12/#:~:text=A%20function%20f%20is%20differentiable,of%20these%20derivatives%20are%20equal.
  https://stackoverflow.com/questions/12915317/how-to-find-derivative-of-a-function-using-c/59128217#59128217
  https://i.stack.imgur.com/3NG5q.jpg

  """


  return (func(a+h)-func(a-h))/(2*h)



def f(x):
  """
  f(x)=x36x4=0
  """
  return x**3 - 6*x-4

xold=x= -4
ep= 10**-6


xnew = xold + 10 * ep # initialize such that |xold - xnew| should initially be greater than ep
iters= 0 

while (ep<abs(xold - xnew)):
  iters+=1
  xold = x
  x = x - f(x)/differentiate(f,x)
  xnew = x

print("Root after",iters,"Iterations is :",x)
````
